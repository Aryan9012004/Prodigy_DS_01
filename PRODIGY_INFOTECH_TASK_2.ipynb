```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.dropna(subset=["Embarked"],inplace=True)
data["Cabin"].fillna("Unknown",inplace=True)
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

**Creating Histogram**

```python
plt.figure(figsize=(6,3))
sns.histplot(data["Age"],kde=True , palette = 'Set2')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of People in age bracket")
plt.show()
```

**Creating Bar Plot**

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex",hue="Sex" , palette = 'Set1')
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived",loc="upper right")
plt.show()
```

**Scatter Plot**

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=data,x="Age",y="Fare", hue="Age")
plt.title("Scatter plot of Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived",loc="upper right")
plt.show()
```
