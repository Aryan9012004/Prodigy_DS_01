**Fetching top 10 country based on data to create plots**

```python
top_10_1960 = data.nlargest(10 , columns=['1970 Population'])
```

```python
top_10_2023 = data.nlargest(10 , columns=['2022 Population'])
```

**Bar Plot**

```python
plt.figure(figsize=(15, 8))
plt.subplot(2,2,1)
sns.barplot(x="1970 Population", y="Country/Territory", data=top_10_1960.head(10),hue='1970 Population' , palette='bright')
plt.title("Top Ten Countries of Total Population (1970)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()


plt.figure(figsize=(15, 10))
plt.subplot(2,2,2)
sns.barplot(x="2022 Population", y="Country/Territory", data=top_10_2023.head(10),hue='2022 Population' , palette='bright')
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
data_1970 = data.nlargest(233 , columns=['1970 Population'])
data_2022 = data.nlargest(233 , columns=['2022 Population'])
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2,2,1)
sns.barplot(x="1970 Population", y="Country/Territory", data=data_1970.tail(10),palette='bright')
plt.title("Top Ten Countries of Total Population (1970)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()


plt.figure(figsize=(15, 10))
plt.subplot(2,2,2)
sns.barplot(x="2022 Population", y="Country/Territory", data=data_2022.tail(10),palette='bright')
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
